### Correlation Matrices

```python
high_corr = ['W_pitching',
             'SO/W_pitching',
             'OPS+_batting',
             'RBI_batting',
             'R/G_batting',
             'R_batting',
             'ERA+_pitching',
             'OPS_batting',
             'SLG_batting',
             'OBP_batting',
             'TB_batting',
             'SV_pitching',
             'SO_pitching',
             'Rdrs_fielding',
             'tSho_pitching']

low_corr = ['W_pitching',
             'R_pitching',
             'RA/G_pitching',
             'ERA_pitching',
             'WHIP_pitching',
             'ER_pitching',
             'FIP_pitching',
             'BF_pitching',
             'H9_pitching',
             'BB9_pitching',
             'LOB_pitching',
             'WP_pitching',
             'HR9_pitching',
             'E_fielding',
             'SH_batting']

years = ['2022',
         '2021',
         '2020',
         '2019',
         '2018',
         '2017']
```

```python
high_correlation = pd.DataFrame(columns = years)
low_correlation = pd.DataFrame(columns = years)
```

```python
for idx, team in enumerate(BB.teams):
    cor = team.corr()
    high_correlation[years[idx]] = cor['W_pitching'].sort_values(ascending = False)[:20]
    low_correlation[years[idx]] = cor['W_pitching'].sort_values()[:20]
    
    team_sub_high = team[high_corr]
    plt.figure(figsize = (20,10))
    plt.title(f'{years[idx]} High Correlation')
    sns.heatmap(team_sub_high.corr(), annot=True)
    
    team_sub_low = team[low_corr]
    plt.figure(figsize = (20,10))
    plt.title(f'{years[idx]} Low Correlation')
    sns.heatmap(team_sub_low.corr(), annot=True)
```
